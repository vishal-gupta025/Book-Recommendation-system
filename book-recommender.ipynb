# Popularity based recommendation system


```python
rating_with_name = ratings.merge(books,on='ISBN')
```

```python
rating_with_name
```

```python
num_rating_df = rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = rating_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()   
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_df.shape
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-L','num_ratings','avg_ratings']]
```

```python
popular_df
```

# Collaborative filtering based recommender system

```python
x = rating_with_name.groupby('User-ID').count()['Book-Rating'] >= 200
exp_users = x[x].index
```

```python
filtered_rating = rating_with_name[rating_with_name['User-ID'].isin(exp_users)]
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating
```

```python
pt = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    distance = similarity_score[index]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1] , reverse=True)[1:11]

    for i in similar_items:
        print(pt.index[i[0]])
    
```

```python
sorted(list(enumerate(similarity_score[0])),key=lambda x:x[1] , reverse=True)[1:11]
```

```python
recommend('The Da Vinci Code')
```
